````
import cv2 
from google.colab.patches import cv2_imshow
img = "./CascadeTabNet/Demo/tbl/tbl/Stock_Open_close.PNG"
result = inference_detector(model, img)
# show_result_pyplot(img, result,('Bordered', 'cell', 'Borderless'), score_thr=0.85)
im = cv2.imread(img)
res_border = []
res_bless = []
res_cell = []
for r in result[0][0]:
    if r[4]>.65:
        res_border.append(r[:4].astype(int))
## for cells
for r in result[0][1]:
    if r[4]>.65:
        r[4] = r[4]*100
        res_cell.append(r.astype(int))
## for borderless
for r in result[0][2]:
    if r[4]>.65:
        res_bless.append(r[:4].astype(int))
res_bless = res_bless + res_border
for i in res_bless:
  start = (i[0],i[1])
  end = (i[2],i[3])
  cv2.rectangle(im, start, end, (255,0,0), 2)
cv2_imshow(im)
````
